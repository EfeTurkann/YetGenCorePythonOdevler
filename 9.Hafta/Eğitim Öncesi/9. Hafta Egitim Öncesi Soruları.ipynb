# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self,tür,renk,ehliyet):
        self.tür = tür
        self.renk = renk
        self.ehliyet = ehliyet
class Bus(Vehicle):
    Bus_seating_capacity = 50
    def __init__(self, tür, renk, ehliyet):
        super().__init__(tür, renk, ehliyet)
        print(f"Bir otobüsün oturma kapasitesi {self.Bus_seating_capacity} yolcudur.")
school_bus = Bus("Otobüs","Sarı","B1")
````
