# Problem 2

2 YetGen lideri ders sırasında canları sıkılınca oyun oynamak isterler. Ancak hangi oyunu oynayacaklarına karar veremezler ve son olarak kendi oyunlarını yazmaya karar verirler.
Liderlerimizin kodlamak istediği oyun kuralları şöyledir:
- Oyunda random olarak 0 ile 100 arasında bir sayı tutulur.
- Kullanıcıdan bilgisayarın tuttuğu sayıyı tahmin etmesi için bir sayı girmesini istenir.
- Eğer kullanıcının girdiği sayı bulunması gereken sayıdan küçükse  tahmininizi yükseltin, büyükse tahmininizi düşürün şeklinde bir çıktı verilir.
- Oyunun gerçeğe yakın olması için bilgisayarın çalışmasını simgelemek için 1 saniye bekleme süresi tanınır.
- Kullanıcı 5 hakkı vardır. 5 hakkı da kullanıp sayıyı bulamazsa oyunu kaybeder.

```python
import random
import time
kullanici_hakki = 5
sabit_kullanici_hakki = kullanici_hakki
print("Sayı tahmin oyununa hoş geldiniz.Bu oyunda sizden" , kullanici_hakki,  "tahmin hakkında sayıyı bilmenizi bekliyoruz.İyi Şanslar!")
random_sayi = random.randint(1,100) #1,100 aralığında bir sayı seçmeye yarar.
#print(random_sayi)

while True:
    if kullanici_hakki == 0:
        print("Hakkınız bitti." "Sayımız;" , random_sayi ,"idi.")
        break
    kullanici_tahmini = int(input("Lütfen tahmininizi bir tam sayı olarak giriniz."))
    if kullanici_tahmini < random_sayi:
        print("Bilgiler işleniyor,lütfen bekleyiniz.")
        time.sleep(1) #Bu aşamaya geldiğinde program 1 saniye çalışmayı durduracak.
        print("Yaklaştınız,daha büyük bir tahminde bulununuz.")
        kullanici_hakki -=1
        print("Kalan hakkınız:" , kullanici_hakki)
    elif kullanici_tahmini > random_sayi:
        print("Bilgiler işleniyor,lütfen bekleyiniz.")
        time.sleep(1) #Bu aşamaya geldiğinde program 1 saniye çalışmayı durduracak.
        print("Yaklaştınız,daha küçük bir tahminde bulununuz.")
        kullanici_hakki -=1
        print("Kalan hakkınız:" , kullanici_hakki)
    elif kullanici_tahmini == random_sayi:
        print("Bilgiler işleniyor,lütfen bekleyiniz.")
        time.sleep(1) #Bu aşamaya geldiğinde program 1 saniye çalışmayı durduracak.
        print("Tebrikler kazandınız !" , (sabit_kullanici_hakki) + 1 - (kullanici_hakki ) , "Tahminde doğru sonuca ulaştınız.")
        kullanici_hakki -=1
        break
```
