# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# YetNot: time kütüphanesini projemize dahil edelim.
import time 
# YetNot: vki_hesaplama isimli kilo ve boy parametresi olan bir fonksiyon yazalım.
def vki_hesaplama(kilo, boy):
  #YetNot : vki isimli bir değişken oluşturup vücut kitle indeksini hesaplayalım.
  vki = kilo/(boy**2)
  # YetNot: vki değişkeninin virgülden sonraki 1 basamağını gösterip döndürelim.
  #virgülden sonra belirli bir basamağını göstermek amacıyla round fonksiyonunu kullanıyoruzz. 
  return round(vki,1)
# YetNot: deger_araligi isimli bir fonksiyon oluşturup vki parametresi atayalım.
def deger_araligi(vki):
  # YetNot: vki değeri 18.5'ten küçükse zayıf return edelim.
  if vki< 18.5:
    return "zayıf"
  # YetNot: vki değeri 24.9'dan küçük ya da eşitse  normal kilolu return edelim.
  elif vki<=24.9:
    return "normal kilolu"
  # YetNot: vki değeri 29.9'dan küçük ya da eşitse  fazla kilolu return edelim.
  elif vki<= 29.9:
    return "fazla kilolu"
  # YetNot: vki değeri 39.9'dan  küçük ya da eşitse  obez return edelim.
  elif vki<=39.9:
    return "obez"
  # YetNot: vki değeri 40'tan büyükse ileri obez return edelim.
  elif vki>=40:
    return "ileri obez"
# YetNot: cikti_olustur isimli fonksiyon oluşturup isim, yas, kilo,boy,vki, deger parametresi yollayalım.
def cikti_olustur(isim, yas, kilo,boy,vki, deger):
  # YetNot: kayit tarihi isimli bir değişken oluşturup kayıt tarihini alalım.
  # ctime tarihi kişilerin anlayabileceği bir biçime dönüştürür. 
  kayit_tarihi= time.ctime(time.time())
  # YetNot: t1 isimli bir değişken oluşturup f string ile ismi yazdıralım. 
  t1 = f"kullanici adi {isim}\n"
  # YetNot: t2 isimli bir değişken oluşturup f string ile yas, boy ve kilo değişkenini yazdıralım. 
  t2 = f"kullanıcının yas, kilo, boy bilgisi {yas}, {kilo},{boy}\n"
  # YetNot: t3 isimli bir değişken oluşturup f string ile vki değerini yazdıralım. 
  t3 = f"kullanıcın vücut kitle indexi {vki}\n"
  # YetNot: t4 isimli bir değişken oluşturup f string ile deger yazdıralım. 
  t4 = f"değer aralığı {deger}\n"
  # YetNot: t5 isimli bir değişken oluşturup f string ile kayıt tarihini yazdıralım. 
  t5 = f"oluşturma tarihi: {kayit_tarihi}\n"

  # YetNot: output isimli bir değişken oluşturup t1,t2,t3,t4 ve t5 değerlerini toplayalım. Sonucu return edelim.
  output= t1 + t2 + t3 + t4 + t5
  return output
# YetNot: main isimli bir fonksiyon oluşturalım.
def main():
    #YetNot: giris_sayisi isimli bir değişken oluşturup kullanıcının kaç kişi gireceği bilgisini alalım.
    giris_sayisi= int(input("kaç kullanıcı verisi gireceksiniz?: "))
    # YetNot: for döngüsü ile giris_sayisi kadar gezelim.
    for i in range(giris_sayisi):
       # YetNot: Kullanıcını kayıt numarasını ekrana bastıralım.
        print(f"kullanıcı kaydı {i+1}")
       # YetNot: isim değişkenini tanımlayarak kullanıcıdan isim  soyisim bilgisini alalım.
        isim = input("isim soyisim: ")
      # YetNot: yas değişkenini tanımlayarak kullanıcıdan yaş bilgisini alalım.
        yas = int(input("yas: "))
      # YetNot: kilo değişkenini tanımlayarak kullanıcıdan kilo bilgisini alalım.
        kilo = float(input("kilo (kg): "))
      # YetNot: boy değişkenini tanımlayarak kullanıcıdan boy bilgisini alalım.
        boy = float(input("boy (m): "))
      # YetNot: vki isimli bir değişken oluşturup vki_hesaplama fonksiyonunu çağıralım.
        vki= vki_hesaplama(kilo, boy)
      # YetNot: deger isimli bir değişken oluşturup deger_araligi fonksiyonunu çağıralım.
        deger = deger_araligi(vki)
      # YetNot: output değişkeni oluşturup cikti_olustur fonksiyonunu çağıralım.
        output = cikti_olustur(isim, yas, kilo,boy,vki, deger)
      # YetNot: with ile dosyayı açıp verileri dosyaya yazdıralım.
        with open("saglık.txt", "a", encoding="utf-8") as f:
          f.write(output)
          f.write("*"*20)
          f.write("\n")
          print ("kullanıcı kaydedildi.")
          
    # YetNot: ekrana veriler 'saglık.txt' dosyasına kaydedildi.\n  yazdıralım.
    print("veriler 'saglık.txt' dosyasına kaydedildi.\n ")
main()


```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 2. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import time
menu = {"elma":0.50,"armut":0.75,"uzum":2.,"cilek":1.75,"muz":2.,"kavun":2.15,"karpuz":1.99}
def display_menu():
    print("-----YETBAR MENU-----")
    for item in list(menu.keys()): # YetNot : for döngüsü ile menunün keylerini gezelim.
        print(f"{item.upper():<10} : {menu[item]} TL") # YetNot : her elemanın harflerini büyütüp fiyatlarını yazdıralım. 
def tek_urun_fiyat_hesaplama(urun, adet):
    assert adet >= 0 , "Ürün adedi 0 ve 0'dan küçük olamaz." #assert python dilinde bir hata ayıklama aracıdır ve bununla daha sonrasında karşılacağımız bir soru şimdiden görmemizi sağlar.
    if urun not in list(menu.keys()): # YetNot : urun menu'nün anahtar değerlerinde yoksa none döndürelim.
        return None
    else: # YetNot : Diğer durumlarda menu içinden urun fiyatını bulup adet ile çarpalım.
        return menu[urun] * adet
def satin_alma():
    display_menu() # YetNot : display_menu fonksiyonunu çağıralım.
    print("Lütfen satın almak istediğiniz ürünü giriniz. Çıkış yapmak için '0'ı tuşlayınız.")
    # YetNot : satin_alinan_urunler ve satin_alma_adedi isimli iki değişken oluşturup boş bir liste atayalım.
    satin_alinan_urunler = []
    satin_alma_adedi = []
    # YetNot : son_fiyat isimli bir değişken oluşturup 0 değerini atayalım.
    son_fiyat = 0
    # YetNot  : while döngüsü ile 1'e basıldığı sürece döngünü devam etmesini sağlayalım.
    while 1:
        eklenecek_urun = input("Hangi üründen eklemek istiyorsunuz? ")
        if eklenecek_urun == "0":
            print("Fiş kaydediliyor.")
            break
        # YetNot : ekleme_adedi isimli bir değişken oluşturup Kaç adet eklemek istiyorsunuz? sorusuna kullanıcıdan yanıt alalım.
        ekleme_adedi = int(input("Kaç adet eklemek istiyorsunuz? ")) 
        # YetNot : fiyat isimli bir değişken oluşturup tek_urun_fiyat_hesaplama fonksiyonunu çağıralım.
        fiyat = tek_urun_fiyat_hesaplama(eklenecek_urun, ekleme_adedi)
        # YetNot : Eğer fiyat None ise ekrana "Yanlış bir giriş yaptınız, ürün eklenemedi." bastıralım.
        if fiyat == None:
            print("Yanlış bir giriş yaptınız, ürün eklenemedi.")
        # YetNot : Diğer durumlarda ekrana f"Satın alma ücreti: {fiyat}.\nSatın almayı onaylamak için 'y' iptal etmek için 'n' giriniz." bastıralım.
        else:
            print(f"Satın alma ücreti: {fiyat}.\nSatın almayı onaylamak için 'y' iptal etmek için 'n' giriniz.")
            # YetNot : check adında bir değişken oluşturup kullanıcıya "Satın almayı onaylıyor musunuz? " diye soralım.
            check = input("Satın almayı onaylıyor musunuz? ")
            # YetNot : check eğer y ise "Satın alım onaylandı." bastıralım
            if check == "y":
                print("Satın alım onaylandı.")
                print("-"*40) # print("a"*400)
                # YetNot : son_fiyat değişkenina fiyat değişkenini ekleyelim. 
                son_fiyat += fiyat
                # YetNot : satin_alinan_urunler listesine eklenecek_urun ekleyelim.
                satin_alinan_urunler.append(eklenecek_urun)
                # YetNot : satin_alma_adedi listesine ekleme_adedi değişkenini ekleyelim. Ara toplamı ekrana bastıralım.
                satin_alma_adedi.append(ekleme_adedi)
                print(f"Ara Toplam: {son_fiyat}")
            # YetNot : check n ise ekrana "Satın alma işlemi iptal edildi." yazdıralım.
            elif check == "n":
                print("Satın alma işlemi iptal edildi.")
                print("-"*40)
    # YetNot : son_fiyat, satin_alinan_urunler, satin_alma_adedi return edelim.
    return son_fiyat, satin_alinan_urunler, satin_alma_adedi
# YetNot  : main fonksiyonunu oluşturalım.
def main():
    print("YetBar'a Hoş Geldiniz!\nMenüyü görüntülemek için '1'e, smoothie yapmak için lütfen '2'e basınız. Çıkmak için '0' giriniz.")
    # YetNot : program isimli bir değişken oluşturup None atayalım.
    program = None
    # YetNot  : while döngüsü ile 1'e basıldığı sürece döngünü devam etmesini sağlayalım.
    while 1:
         # YetNot : program isimli bir değişken oluşturup tercihi soralım.
        program = int(input("Tercihiniz: "))
        # YetNot : program 0'dan küçükse ya da 3'ten büyükse ekran ""Yanlış bir tercih yaptınız. Lütfen tekrar deneyiniz."" bastıralım.
        if program < 0 or program > 2:
            print("Yanlış bir tercih yaptınız. Lütfen tekrar deneyiniz.")
        # YetNot : program 0'a eşitse ekrana "Çıkış yapılıyor...\nSağlıklı günler dileriz." yazdıralım.
        elif program == 0:
            print("Çıkış yapılıyor...\nSağlıklı günler dileriz.")
            break
        # YetNot : program 1'e eşitse display_menu fonksiyonunu çağıralım.
        elif program == 1:
            display_menu()
       # YetNot : program 2 ise satin_alma fonksiyonunu çağıralım.
        elif program == 2:
            satin_alma_ucreti, satin_alinan_urunler, satin_alma_adedi = satin_alma()
            # YetNot : with ile dosyayı açıp verileri dosyaya yazdıralım.
            with open("kasa.txt", "a", encoding="utf-8") as f:
                f.write(str(time.ctime(time.time())))
                f.write("\n")
                f.write(','.join(map(str, satin_alinan_urunler)))
                f.write("\n")
                f.write(','.join(map(str, satin_alma_adedi)))
                f.write("\n")
                f.write(f"Toplam: {satin_alma_ucreti}")
                f.write("\n")
                f.write("*"*20)
                f.write("\n")
            # YetNot : cikis_kontrol isimli bir değişken oluşturup "Başka bir işlem yapmak istiyor musunuz? (evet-hayır): " diye soralım.
            cikis_kontrol = input("Başka bir işlem yapmak istiyor musunuz? (evet-hayır): ")
            # YetNot : cikis_kontrol evet ise pass edelim.
            if cikis_kontrol == "evet":
                pass
            # YetNot : cikis_kontrol evet ise pass edelim. Ekrana \n\nTEŞEKKÜRLER yazdıralım ve döngüden çıkalım.
            else:
                print("\n\nTEŞEKKÜRLER")
                break
main()
```
