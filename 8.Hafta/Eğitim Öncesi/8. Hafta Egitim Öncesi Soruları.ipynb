# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def add_product():
    ürün = input("Lütfen eklemek istediğiniz ürünü giriniz.")
    fiyat = input("Lütfen eklemek istediğiniz ürünün fiyatını giriniz.")
    with open("ürünler.txt" , "a" , encoding="utf-8") as f:
        f.write(f"Ürün İsmi:{ürün} Ürün Fiyatı:{fiyat}\n")
add_product()
````

````python
with open("ürünler.txt" , "r" , encoding="utf-8") as f:
    print(f.read())
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
def degistir(dosya_ismi,eski_kelime,yeni_kelime):
    with open(dosya_ismi , "r+" , encoding="utf-8") as f:
        liste = f.read()
        yeni_liste = liste.replace(eski_kelime,yeni_kelime)
        f.seek(0)
        f.write(yeni_liste)

degistir("ürünler.txt","6000","7750")
degistir("ürünler.txt","S10","S9")
````

````python
with open("ürünler.txt" , "r" , encoding="utf-8") as f:
    print(f.read())
````

Alt sınıf oluşturma:

````python
class Vehicle:
    def __init__(self,name,speed,mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
    def show_info(self):
        print(f"Araç adı:{self.name} Araç Hızı:{self.speed} Mesafe:{self.mileage}")
class Bus(Vehicle):
    pass
bus1 = Bus("School Volvo",180,12)
print(bus1.name)  
print(bus1.speed) 
print(bus1.mileage)
(bus1.show_info()) 
````
